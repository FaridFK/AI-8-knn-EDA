```python
#KLASIFIKASI MENGGUNAKAN KNN

#Impor library yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import KFold
import seaborn as sns
import itertools


#Mengambil data dari file csv untuk data latih
dataset=pd.read_csv('Social_Network_Ads.csv')
df=pd.DataFrame(dataset,columns=["User ID","Gender","Age","EstimatedSalary","Purchased"])
x = dataset.iloc[:, [2, 3]].values
y = dataset.iloc[:, -1].values

from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.25, random_state = 0)

#Menampilkan data
print(dataset)
print('----------------------------------------------')

#Menampilkan deskripsi data
print(dataset.describe())
print('----------------------------------------------')

#Menampilkan data dengan nilai nol
print(dataset.isnull().sum())
print('----------------------------------------------')

#Menampilkan tipe data yang digunakan
print(dataset.info())
print('----------------------------------------------')

# Visualisasi dataset
sns.countplot(x='Purchased',data=dataset)
plt.show()

# Menghitung data pada variabel Age
sns.countplot(dataset['Age'])
plt.show()

# Menghitung data pada variabel Estimated Salary
sns.countplot(dataset['EstimatedSalary'])
plt.show()

#Melihat lebih detil data variabel Age
plt.hist(dataset['Age'],bins=10,edgecolor='black')
plt.xlabel('Range')
plt.ylabel("Age")
plt.title('Closure look at Age')
plt.show()

#Mencari outlier di variabel Age
dt=dataset.loc[dataset['Age']<25]
print(dt.shape[0])
print("Before concatenation rows :" ,dt.shape[0])
dt=dt.append(dt)
print("After concatenation rows :" ,dt.shape[0])

#Melihat lebih detil data di variabel Estimated Salary
plt.hist(dataset['EstimatedSalary'],bins=10,edgecolor='black')
plt.xlabel('Range')
plt.ylabel("Salary")
plt.title('Closure look at Estimated Salary')
plt.show()

#Mencari outlier di variabel Age
dt=dataset.loc[dataset['EstimatedSalary']>25000]
print(dt.shape[0])
print("Before concatenation rows :" ,dt.shape[0])
dt=dt.append(dt)
print("After concatenation rows :" ,dt.shape[0])
data1=dataset.copy()
data1=data1.loc[data1['Age']<=25]
data1=data1.loc[data1['EstimatedSalary']>=25000]
data1.shape[0]

sns.pairplot(data=dataset,hue='Purchased',diag_kind='kde')
plt.show()

#Memanggil metode KNN
knn=KNeighborsClassifier(n_neighbors=6, 
                         weights='uniform',
                         algorithm='auto',
                         metric='euclidean')

#Fit model dengan data latih dan target
knn.fit(x_train,y_train)

#Melakukan prediksi
y_pred=knn.predict(x_val)
print((y_val).T)
print(y_pred)
print(classification_report(y_val,y_pred))

#Visualisasi Data
#plot Data Latih
absis=[x_train[:,0]]
ordinat=[x_train[:,1]]
target=[y_train]
plt.scatter(absis, ordinat, alpha=1, c=target, cmap='flag', marker='+')

#plot Data Uji
absis=[x_val[:,0]]
ordinat=[x_val[:,1]]
target=[y_val]
plt.scatter(absis, ordinat, alpha=2, c=target, cmap='flag', marker='x')

#Label
plt.xlabel("Age")
plt.ylabel("Salary)")
plt.title("KNN Classification")
plt.grid()

#Memprediksi data baru
data=[[20,150000]]
prediksi=knn.predict(data)
print(prediksi)
```
